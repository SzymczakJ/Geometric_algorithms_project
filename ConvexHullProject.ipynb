# Time comparison

```python
from time import perf_counter
import pandas as pd
```

```python
from algorithms_without_visualisation.chan_convex_hull_algorithm import chan_convex_hull as fast_chan
from algorithms_without_visualisation.divide_and_conquer_convex_hull_algorithm import divide_and_conquer as fast_d_and_c
from algorithms_without_visualisation.graham_convex_hull_algorithm import graham_convex_hull as fast_graham
from algorithms_without_visualisation.incremental_convex_hull_algorithm import incremental_convex_hull as fast_incremental
from algorithms_without_visualisation.jarvis_convex_hull_algorithm import jarvis_convex_hull as fast_jarvis
from algorithms_without_visualisation.quickhull_convex_hull_algorithm import quickhull_convex_hull as fast_quickhull
from algorithms_without_visualisation.upper_lower_convex_hull_algorithm import upper_lower_convex_hull as fast_upper_lower
```

```python
def count_time(algorithm, data):
    start = perf_counter()
    algorithm(data)
    end = perf_counter()
    time = end - start
    return time
```

```python
def time_comparison(chan, d_and_c, graham, incremental, jarvis, quickhull, upper_lower):
    columns = ["Chan", "Divide and conquer", "Graham", "Incremental", "Jarvis",
               "Quickhull", "Upper-lower"]
    data = []
    for rd in random_data:
#         chan_time = count_time(chan, rd)
        d_and_c_time = count_time(d_and_c, rd)
#         graham_time = count_time(graham, rd)
#         incremental_time = count_time(incremental, rd)
#         jarvis_time = count_time(jarvis, rd)
#         quickhull_time = count_time(quickhull, rd)
        upper_lower_time = count_time(upper_lower, rd)
        chan_time = 0
#         d_and_c_time = 0
        graham_time = 0
        incremental_time = 0
        jarvis_time = 0
        quickhull_time = 0
        upper_lower_time = 0
        data.append((chan_time, d_and_c_time, graham_time, incremental_time,
                     jarvis_time, quickhull_time, upper_lower_time))
    return pd.DataFrame(data=data, columns=columns)
```

```python
time_comparison_df = time_comparison(fast_chan, fast_d_and_c, fast_graham, fast_incremental,
                                 fast_jarvis, fast_quickhull, fast_upper_lower)
time_comparison_df
```
